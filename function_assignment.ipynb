1:-. What is the difference between a function and a method in Python?

Ans:-In Python, both functions and methods are blocks of reusable code designed to perform specific tasks, but their key distinction lies in their relationship with objects and classes.
Function:
A function is a standalone block of code that is not associated with any particular object or class.
It is defined using the def keyword and can be called independently by its name from anywhere in the program.
Functions can take arguments and return values.
Example:
Python

def greet(name):

    return f"Hello, {name}!"

message = greet("Alice")

print(message)

Method:

A method is a function that is defined within a class and is associated with an object (an instance of that class) or the class itself.
Methods are invoked using dot notation on an object or class.
They implicitly receive the instance (conventionally named self) or the class (conventionally named cls) as their first argument, allowing them to access and manipulate the object's or class's attributes and other methods.

Example:


class Dog:

    def __init__(self, name):
        self.name = name

    def bark(self):  # This is a method
        return f"{self.name} says Woof!"

my_dog = Dog("Buddy")

print(my_dog.bark())


2:- Explain the concept of function arguments and parameters in Python.

Ans:-In Python, the terms "parameters" and "arguments" are distinct concepts related to functions:

Parameters:

Parameters are placeholders defined in the function's signature (its definition). They act as variables that will receive values when the function is called. Parameters are specified within the parentheses of the def statement.



def greet(name):  # 'name' is a parameter

    print(f"Hello, {name}!")

Arguments:

Arguments are the actual values passed to a function when it is called or invoked. These values are assigned to the corresponding parameters defined in the function's signature.


greet("Alice")  # "Alice" is an argument

Relationship:

When greet("Alice") is called, the string "Alice" (the argument) is passed to the greet function. Inside the function, this value is assigned to the name parameter, allowing the function to use "Alice" in its operations.

In summary:

Parameters

are variables in the function definition, acting as placeholders for expected inputs.

Arguments

are the actual values provided when the function is called, which fill those placeholders.

4:- What is the purpose of the `return` statement in a Python function?

Ans:-The return statement in a Python function serves the purpose of ending the function's execution and sending a value back to the caller. It allows functions to produce results and communicate those results to other parts of the program.

Here's a more detailed explanation:

Exiting the Function:

When a return statement is encountered, the function immediately stops executing, even if there are more lines of code after it.

Returning a Value:

The return statement can optionally be followed by an expression (e.g., a variable, a calculation, etc.). This expression's value is then returned to the caller.

Communicating Results:

The returned value can be used by the caller for further processing or stored in a variable.

Implicit Return of None:

If a return statement is used without an expression, or if the function has no return statement at all, it implicitly returns None.

Example:


    def add_numbers(x, y):
        sum = x + y
        return sum  # Returns the calculated sum

6:- Explain the concept of generators in Python and how they are defined.

Ans:- A generator is a function that produces a sequence of values one at a time, rather than computing and storing the entire sequence in memory at once. This makes generators memory-efficient, especially when dealing with large datasets or infinite sequences. Generators are defined like regular functions, but they use the yield keyword instead of return.
def my_generator(n):

Example:-

    """Generates numbers from 0 up to n-1."""
    value = 0
    while value < n:
        yield value
        value += 1

# Create a generator object
my_gen = my_generator(5)

# Iterate through the generator
for number in my_gen:

    print(number)

# Output:
# 0
# 1
# 2
# 3
# 4

7:- What are the advantages of using generators over regular functions?

Ans:-Generators in Python offer several advantages over regular functions, especially when working with large datasets or streams of data. Here's a

breakdown:

**Advantages of Generators

Memory Efficiency

Generators produce one item at a time using lazy evaluation.

They don't store the entire sequence in memory like lists.

Performance

Ideal for large or infinite sequences.

Faster startup time since they yield items one-by-one.

Convenient Syntax

Easier to write and read than managing iterator classes manually.

State Preservation

Automatically remember the state between yield calls without extra code.

Useful in Pipelines

Can be chained together for stream processing, like Unix pipes.

 Example: Generator vs Regular Function

 Regular Function (Returns List)


def get_squares_list(n):

    result = []
    for i in range(n):
        result.append(i * i)
    return result

print(get_squares_list(5))
# Output: [0, 1, 4, 9, 16]
Builds and returns the entire list in memory.

** Generator Function (Uses yield)


def get_squares_gen(n):

    for i in range(n):
        yield i * i

for square in get_squares_gen(5):

    print(square)
# Output: 0 1 4 9 16 (one at a time)
Generates one item at a time.

Memory-efficient — especially useful for large n.

1. With map():

9:- Explain the purpose and usage of the `map()` function in Python.

Ans:-The map() function in Python is a built-in higher-order function used to apply a specified function to each item in an iterable (such as a list, tuple, or set). It provides a concise and efficient way to transform data without explicitly writing a for loop.

Purpose:

The primary purpose of map() is to perform an operation on every element of an iterable, generating a new iterable with the results. This is

particularly useful for:

Data Transformation:

Applying a consistent change to all elements in a collection (e.g., converting strings to uppercase, squaring numbers).

Code Readability:

Offering a more functional and often more readable alternative to explicit loops for element-wise operations.
Efficiency:

For large datasets, map() can sometimes be more efficient than a for loop, especially when implemented in C (as built-in functions often are).

Usage:

The syntax for map() is:



map(function, iterable, ...)

function: The function to be applied to each item of the iterable(s). This can be a named function, a lambda function, or even a built-in function like int or str.

iterable: One or more iterable objects whose elements will be passed to the function. If multiple iterables are provided, the function must accept a corresponding number of arguments, and map() will iterate until the shortest iterable is exhausted.

Return Value:

map() returns an iterator (a map object), not a list directly. This iterator can then be converted into a list, tuple, set, or other desired collection type using the respective constructor (e.g., list(), tuple()).

Example:


# Applying a function to a single iterable
numbers = [1, 2, 3, 4]

squared_numbers = list(map(lambda x: x * x, numbers))

print(squared_numbers)

# Applying a function to multiple iterables
list1 = [1, 2, 3]

list2 = [10, 20, 30]

sum_lists = list(map(lambda x, y: x + y, list1, list2))

print(sum_lists)

10:-What is the difference between `map()`, `reduce()`, and `filter()` functions in Python?

Ans:- comparison of map(), reduce(), and filter() in Python — all of which are part of the functional programming toolkit.

map(function, iterable)

Purpose: Applies a function to each item in an iterable.

Returns: A map object (which can be converted to list, etc.)

Use case: When you want to transform all elements.

Example:

nums = [1, 2, 3, 4]

squared = map(lambda x: x ** 2, nums)

print(list(squared))  # Output: [1, 4, 9, 16]

 filter(function, iterable)

Purpose: Applies a function to each item and keeps only those where the function returns True.


Returns: A filter object.

Use case: When you want to filter elements based on a condition.

Example:

nums = [1, 2, 3, 4, 5]

evens = filter(lambda x: x % 2 == 0, nums)

print(list(evens))  # Output: [2, 4]

 reduce(function, iterable)

Purpose: Applies a function cumulatively to reduce the iterable to a single value.

Returns: A single result.

Use case: When you want to aggregate or accumulate results (e.g., sum, product).

Note: reduce() is part of the functools module.

 Example:

from functools import reduce

nums = [1, 2, 3, 4]

product = reduce(lambda x, y: x * y, nums)

print(product)

```python
 """ 1. Write a Python function that takes a list of numbers as input and returns the sum of all even numbers in
the list. """

def sum_even_numbers(numbers):
    even_sum = 0
    for num in numbers:
        if num % 2 == 0:
            even_sum += num
    return even_sum

# Example list
my_list = [1, 2, 3, 4, 5, 6]

# Call the function and print the result
result = sum_even_numbers(my_list)
print("Sum of even numbers:", result)

```

```python
#4. Write a Python function that checks if a given number is prime or not from 1 to 200.
def is_prime(n):
    if n <= 1:
        return False
    for i in range(2, int(n**0.5) + 1):
        if n % i == 0:
            return False
    return True

# Check numbers from 1 to 200 and print if they are prime
for number in range(1, 201):
    if is_prime(number):
        print(f"{number} is a prime number")
```

```python
#5. Create an iterator class in Python that generates the Fibonacci sequence up to a specified number of term
class FibonacciIterator:
    def __init__(self, max_terms):
        self.max_terms = max_terms
        self.count = 0
        self.a, self.b = 0, 1

    def __iter__(self):
        return self

    def __next__(self):
        if self.count >= self.max_terms:
            raise StopIteration
        fib = self.a
        self.a, self.b = self.b, self.a + self.b
        self.count += 1
        return fib

# Example usage
fib = FibonacciIterator(10)  # Generate first 10 Fibonacci numbers
for num in fib:
    print(num)
```

```python
#6. Write a generator function in Python that yields the powers of 2 up to a given exponent.
def powers_of_two(max_exponent):
    for i in range(max_exponent + 1):
        yield 2 ** i

# Example usage
print("Powers of 2 up to exponent 5:")
for power in powers_of_two(5):
    print(power)
```

```python
#7. Implement a generator function that reads a file line by line and yields each line as a string.
def read_file_line_by_line(file_path):
    """
    Reads a file line by line and yields each line as a string.

    Args:
        file_path (str): The path to the file to be read.

    Yields:
        str: Each line of the file as a string, with leading/trailing whitespace removed.
    """
    try:
        with open(file_path, 'r') as file:
            for line in file:
                yield line.strip()  # Yield each line, removing leading/trailing whitespace
    except FileNotFoundError:
        print(f"Error: The file '{file_path}' was not found.")
    except Exception as e:
        print(f"An error occurred: {e}")

# Example Usage:
# Create a dummy file for demonstration
with open("sample.txt", "w") as f:
    f.write("This is line 1.\n")
    f.write("This is line 2.\n")
    f.write("And this is line 3.\n")

# Use the generator function to read and print lines
for line in read_file_line_by_line("sample.txt"):
    print(line)

# Example with a non-existent file
print("\nAttempting to read a non-existent file:")
for line in read_file_line_by_line("non_existent_file.txt"):
    print(line)

```

```python
#9. Write a Python program that uses `map()` to convert a list of temperatures from Celsius to Fahrenheit.
# List of temperatures in Celsius
celsius_temps = [0, 20, 37, 100]

# Conversion function: Fahrenheit = (Celsius * 9/5) + 32
fahrenheit_temps = list(map(lambda c: (c * 9/5) + 32, celsius_temps))

print("Celsius temperatures:", celsius_temps)
print("Fahrenheit temperatures:", fahrenheit_temps)
```

```python
#10. Create a Python program that uses `filter()` to remove all the vowels from a given string.
def remove_vowels(text):
    vowels = "aeiouAEIOU"
    # filter out characters that are vowels
    filtered_chars = filter(lambda char: char not in vowels, text)
    return ''.join(filtered_chars)

# Example usage
input_str = "Hello, how are you?"
result = remove_vowels(input_str)
print("Original string:", input_str)
print("String without vowels:", result)
```
